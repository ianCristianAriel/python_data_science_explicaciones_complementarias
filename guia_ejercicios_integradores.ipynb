#### Ejercicio 1: Emulación de formato tabular

En las clases anteriores, utilizamos listas para representar columnas y diccionarios para enlazarlas bajo una misma estructura. En este ejercicio, construirás el "molde" que transformará un diccionario de Python en una estructura tabular inteligente.

Tarea:

a. Crea una clase llamada DataTable que reciba un diccionario de transacciones y contenga:

* Atributo shape: No se recibe como parámetro. Debe implementar una lógica que verifique si todas las listas del diccionario tienen la misma longitud. Si no coinciden, debe lanzar una excepción de tipo ValueError avisando del error de integridad.

* Atributo data: Almacenará el diccionario recibido solo si se cumple la validación de shape.

````python
# Completar
````

##### Solución

````python
class DataTable:

    def __init__(self, data: dict):
        # Validar que todas las listas tengan la misma longitud
        longitudes = [len(columna) for columna in data.values()]

        if len(set(longitudes)) != 1:
            raise ValueError("Error de integridad: las columnas no tienen la misma longitud.")

        # Si pasa la validación, guardamos los datos
        self.data = data

        # Definimos el shape como (filas, columnas)
        self.shape = (longitudes[0], len(data))
````

b. Redefine la clase añadiendo capacidad de consulta precisa:

* Método obtener_columna(nombre): Debe retornar la lista completa asociada a la clave proporcionada (ej: "producto").

* Método leer_celda(fila, columna): Debe recibir el índice de la fila (int) y el nombre de la columna (str) para retornar el valor exacto en esa posición (ej: fila 4 de "monto" $\rightarrow$ 350.00).

````python
# Completar
````

##### Solución

````python
class DataTable:

    def __init__(self, data: dict):
        # Validación de integridad (shape)
        longitudes = [len(columna) for columna in data.values()]

        if len(set(longitudes)) != 1:
            raise ValueError("Error de integridad: las columnas no tienen la misma longitud.")

        self.data = data
        self.shape = (longitudes[0], len(data))

    def obtener_columna(self, nombre: str):
        """
        Retorna la lista completa asociada a una columna.
        """
        return self.data[nombre]

    def leer_celda(self, fila: int, columna: str):
        """
        Retorna el valor específico en una posición (fila, columna).
        """
        return self.data[columna][fila]
````

#### Ejercicio 2: POO como asistente estandarizado de tratamiento de datos

Para que nuestro código sea profesional, podemos encapsular lógicas encargadas de aplicar tratamientos específicos sobre las columnas. b. Para que el transformador tenga sentido, ademas de recibir un atributo debe implementar ciertas funciones de limpieza. Para ello redefinir la clase data_to_process ańadiendole estos metodos:

* Método .normalizer(nombre_columna): Debe acceder a la columna de strings indicada (ej: "producto"), eliminar espacios en blanco al inicio y final, y asegurar que solo la primera letra sea mayúscula (formato Capital).

* Método .deduplicate(columna_id): Debe limpiar la tabla de registros repetidos basándose en un identificador único (ej: "id_transaccion"). Si detecta que un ID (como T001) aparece más de una vez, debe conservar solo la primera aparición y eliminar las filas sobrantes en todas las columnas para mantener la consistencia.

````python
# Completar
````

##### Solución

````python
class DataCleaner:

    def __init__(self, data: dict):
        self.data_to_process = data

    def normalizer(self, nombre_columna: str):
        """
        Limpia una columna de texto:
        - Elimina espacios en blanco al inicio y final
        - Convierte a formato Capital (primera letra mayúscula)
        """
        columna = self.data_to_process[nombre_columna]

        self.data_to_process[nombre_columna] = [
            valor.strip().capitalize() if isinstance(valor, str) else valor
            for valor in columna
        ]

    def deduplicate(self, columna_id: str):
        """
        Elimina registros duplicados basándose en una columna ID.
        Conserva la primera aparición y mantiene consistencia en todas las columnas.
        """
        data = self.data_to_process

        vistos = set()
        indices_validos = []

        # Identificar filas únicas
        for i, valor_id in enumerate(data[columna_id]):
            if valor_id not in vistos:
                vistos.add(valor_id)
                indices_validos.append(i)

        # Reconstruir el diccionario filtrando por índices válidos
        for columna in data.keys():
            data[columna] = [data[columna][i] for i in indices_validos]
````

En este ejercicio, aplicaremos de manera manual, tratamientos estadísticos comunes sobre la columna "monto" para prepararla para [futuras librerías](https://scikit-learn.org/stable/api/sklearn.preprocessing.html).

a. Utiliza NumPy para calcular la media ($\mu$) y la desviación estándar ($\sigma$) de la columna "monto".
